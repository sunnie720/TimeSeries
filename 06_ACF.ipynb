## Autocovariance function

100 time series datapoints with no special pattern 

```r
purely_random_process=ts(rnorm(100))
```

생성한 시계열 데이터의 acutocovariance coeffient 확인
- lag(시차) 별로 autocovariance coeffient 계산값 반환

```r
(acf(purely_random_process, type='covariance'))
```

## Autocorrelation function 

acf() 명령어의 'type' 인자값을 입력하지 않으면 default가 'correlation'


acutocorrelation coeffient 확인
- lag(시차) 별로 acutocorrelation coeffient 계산값 반환
- 처음 timestep 0에서의 1 제외하고는 상관관계 찾기 어려움 => stationarity 정상성

```r
acf(purely_random_process, main='Correlogram of purely random process')
```

```r
(acf(purely_random_process, main='Correlogram of purely random process'))
```
