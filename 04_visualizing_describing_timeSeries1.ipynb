# Plot time series data (시계열 데이터 시각화하기)

````r
require(astsa)
````

````r
help(astsa)
````

````r
help(jj) 
# jj - Johnson and Johnson Quarterly Earnings Per Share (Time Series data)
````

````r
plot(jj, type='o', main='Johnson&Johnson quarterly earnings per share', ylab='Earnings', xlab='Years')
````

- There are fluctuations, the seasonal variations on the trend
- 추후 stationary principle(정상성)에 위배되는 지 확인 
  - transitional affect
  - different variation in different parts of time series

````r
help(flu)
# Monthly pneumonia and influenza deaths per 10,000 people in the United States for 11 years, 1968 to 1978.
# time series data
````

````r
plot(flu, main='Monthly Pneumonia and Influenza Deaths in US',
ylab='Number of Deaths per 10,000 people', xlab='Months')
````

- 년도별로 peak를 찍고 감소하는 seasonality가 보임
- 장기적으로는 우하향 추세

````r
help(globtemp)
# Global mean land-ocean temperature deviations (from 1951-1980 average), measured in degrees centigrade, for the years 1880-2015.
# time series data
````

````r
plot(globtemp, main='Global mean land-ocean deviations from average temperature of 1951-1980',
ylab='Temperature deviations', type='o', xlab='Years')
````

- 장기적으로 우상향 추세

````r
help(globtempl)
# Global mean [land only] temperature deviations (from 1951-1980 average), measured in degrees centigrade, for the years 1880-2015.
# time series data
````

````r
plot(globtempl, main='Global mean land deviations from average temperature of 1951-1980',
ylab='Temperature deviations', type='o', xlab='Years')
````

- land-ocean 보다 land only로 봤을 때 더 확실한 우상향 추세

````r
help(star)
# The magnitude of a star taken at midnight for 600 consecutive days.
# time series data
````

````r
plot(star, main='The magnitude of a star taken at midnight for 600 consecutive days',
ylab='Magnitude', xlab='Days')
````
