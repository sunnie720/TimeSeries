# Codes for Histogram
- hist()

```r
small.size.dataset=c(91,49,76,112,97,42,70, 100, 8, 112, 95, 90, 78, 62, 56, 94, 65, 58, 109, 70, 109, 91, 71, 76, 68, 62, 134, 57, 83, 66)
```

```r
# 기본 히스토그램
hist(small.size.dataset)  # y-axis : frequency
```

```r
# Changing x-label : using 'xlab'
hist(small.size.dataset, xlab='My data points')
```

```r
# changing figure title : using 'main'
hist(small.size.dataset, xlab='My data points', main='Histogram of my data')
```

```r
# changing y-axis to probability : using 'freq=F'
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=F) # freq=False
```

```r
# changing column color : using 'col'
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=F, col='green')
```

```r
# adding density lines : using 'lines(density(dataset))'
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=F, col='green')
lines(density(small.size.dataset))
```

```r
# changing formats of density lines 
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=F, col='green')
lines(density(small.size.dataset), col='red', lwd=5)
```

```r
# changing bin width : using 'breaks'
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=F, col='green', breaks=10)
```

```r
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=F, col='green', breaks=10)
lines(density(small.size.dataset), col='red', lwd=5)
```

# Codes for Scatterplot

```r
set.seed(2016)  
```

```r
Test_1_scores=round(rnorm(50, 78, 10)) # rnorm(N, mean, standard deviation) 정규분포
Test_2_scores=round(rnorm(50, 78, 14))
```

```r
print(Test_1_scores)
```

```r
print(Test_2_scores)
```

```r
# 기본 산점도
plot(Test_2_scores~Test_1_scores)
```

```r
# adding figure title, changing x and y label
plot(Test_2_scores~Test_1_scores, main='Test scores for two exams (50 students)', xlab='scores at test 1', ylab='scores at test 2')
```

```r
# changing dot color : using 'col'
plot(Test_2_scores~Test_1_scores, main='Test scores for two exams (50 students)', xlab='Test_1_scores', ylab='Test 2 scores', col='blue')
```
